# Handling Null Values

```python
titanic_df.isnull()
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.isnull().mean()*100
```

```python
titanic_df.isnull().any()
```

#### From the above mean we will conclude that 19% data is not present in Sex and also 77% data is not present in cabin

```python
## We are seperating the columns which have more than 35% missing values hence

titanic_drop_col= titanic_df.isnull().sum()[titanic_df.isnull().sum()>(35/100 * titanic_df.shape[0])]
titanic_drop_col
```

```python
titanic_drop_col.index
```

```python
titanic_df.drop(titanic_drop_col.index,axis=1,inplace=True)
## Here we had deleted the Cabin attribute since it is less useful and has much missing values
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.fillna(titanic_df.mean(),inplace=True)
titanic_df.isnull().sum()

#### since embarked has strings in it so we can fill it with next command as shown in next line
## and also we are unabe to calculate strings mean(). so we use to fill the columns as a string.
```

```python
titanic_df['Embarked'].describe()

# Even though we have 3 unique values in Embarked it has more S values and only 2 values are missing we can add S string to it.
```

```python
titanic_df['Embarked'].fillna('S',inplace=True)
```

```python
titanic_df.isnull().sum()
```

```python
### since we have no further null values we will go to next step....
titanic_df.corr()
```

```python
titanic_df['Familysize']=titanic_df['SibSp']+titanic_df['Parch']
titanic_df.describe()
```

```python
titanic_df.drop(['SibSp','Parch'],inplace=True,axis=1)
```

```python
titanic_df.describe()
```

```python
titanic_df.corr()
```

```python
sns.heatmap(titanic_df.corr(),annot=True)
```

> From the above Heat map we will conclude that The correction is good for the both pclass and Fare.......<br><br>And also for the  Pclass and Age <br> <br>as well as for the fare and survived has less corr().



```python
titanic_df['Alone']=[0 if titanic_df['Familysize'][i]>0 else 1 for i in titanic_df.index]
titanic_df
```

```python
titanic_df.groupby(['Alone'])['Survived'].mean()
```

##### From the above analysis we will say there are less chances of Surviving for the people who had only 1 person ('Alone')<br><br>And the Reason might be the person who are travelling belongs to Rich class or may be prioritised over their

```python
# if person is not alone the chances of ticket price is very high
sns.heatmap(titanic_df[['Alone','Fare']].corr(),annot=True)
```

```python
titanic_df['Sex']=[0 if titanic_df['Sex'][i]=='male' else 1 for i in titanic_df.index]
titanic_df.groupby(['Sex'])['Survived'].mean()
```

> It shows that Female passengers has mmore changes to survival Rate When commpared to male where as:<br>
74% of females are survived <br>
18 % males are survived

```python
titanic_df.groupby(['Embarked'])['Survived'].mean()
```

> it shows that that The Embarked port with 'C' class has More when compared to 'Q' and 'S'
